## Distribution of Age

```python
plt.subplots(figsize=(10,10))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
```

## Gender Distribution in the Data Set

```python
plt.subplots(figsize=(10,10))
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
```

## Box Plots 

```python
plt.subplots(figsize=(10,10))
sns.boxplot(x='Category', y='Purchase Amount (USD)', data=df)
plt.title('Purchase Amount Distribution by Category')

```

## EDA

```python
#Top 10 Geographical Distribution of Customers

top_locations = df['Location'].value_counts().nlargest(10).index

plt.subplots(figsize=(15, 8))
sns.countplot(x='Location', data=df, order=top_locations)
plt.title('Top 10 Geographical Distribution of Customers')
plt.xticks(rotation=45)
plt.show()
```

```python
#Frequency of Purchases

sns.countplot(x='Frequency of Purchases', data=df)
plt.title('Frequency of Purchases Distribution')
plt.xticks(rotation=45)
```

```python
top_locations = df['Location'].value_counts().nlargest(20).index

plt.subplots(figsize=(15, 8))
sns.barplot(x='Location', y='Purchase Amount (USD)', data=df[df['Location'].isin(top_locations)])
plt.title('Top 20 Purchase Patterns in Different States')
plt.xticks(rotation=45)
plt.show()
```

```python
#Top 10 Color Popularity

top_colors = ['Olive', 'Yellow', 'Silver', 'Teal', 'Green', 'Black', 'Cyan', 'Violet', 'Gray', 'Maroon']

color_palette = {'Olive': 'olive', 'Yellow': 'yellow', 'Silver': 'silver', 'Teal': 'teal',
                 'Green': 'green', 'Black': 'black', 'Cyan': 'cyan', 'Violet': 'violet',
                 'Gray': 'gray', 'Maroon': 'maroon'}

colors = [color_palette[color] for color in top_colors]

plt.subplots(figsize=(12, 6))
sns.countplot(x='Color', data=df[df['Color'].isin(top_colors)], order=top_colors, palette=colors)
plt.title('Top 10 Color Popularity')
plt.xticks(rotation=45)
plt.show()
```

```python
#Size Popularity

plt.subplots(figsize=(10,10))
sns.countplot(x='Size', data=df, order=df['Size'].value_counts().index)
plt.title('Size Popularity')

plt.tight_layout()
plt.show()
```

```python
correlation_matrix = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.countplot(x='Location', hue='Size', data=df)
plt.title('Distribution of Sizes Across Different Locations')
plt.xticks(rotation=45)
plt.legend(title='Size', title_fontsize='12')
plt.show()
```

```python
# Plots new_england_states, west_coast_states, texas_colorado_states

new_england_states = ['Connecticut', 'Maine', 'Massachusetts', 'New Hampshire', 'Rhode Island', 'Vermont']
west_coast_states = ['California', 'Oregon', 'Washington']
texas_colorado_states = ['Texas', 'Colorado']

selected_states = new_england_states + west_coast_states + texas_colorado_states
filtered_df = df[df['Location'].isin(selected_states)]


plt.figure(figsize=(15, 8))
sns.countplot(x='Location', hue='Size', data=filtered_df)
plt.title('Distribution of Sizes in Selected States')
plt.xticks(rotation=45)
plt.legend(title='Size', title_fontsize='12')
plt.show()
```

```python
## Top 10 Selling Items Across Genders
top_selling_items = df['Item Purchased'].value_counts().nlargest(10).index

filtered_df = df[df['Item Purchased'].isin(top_selling_items)]

plt.figure(figsize=(12, 6))
sns.countplot(x='Item Purchased', hue='Gender', data=filtered_df)
plt.title('Top 10 Selling Items Across Genders')
plt.xticks(rotation=45)
plt.legend(title='Gender', title_fontsize='12')
plt.show()
```

```python
plt.figure(figsize=(10, 10))

normalized_df = df.groupby(['Shipping Type', 'Gender']).size().unstack().apply(lambda x: x / x.sum(), axis=1)

normalized_df.plot(kind='bar', stacked=True)
plt.title('Distribution of Shipping Types Across Genders (Normalized)')
plt.legend(title='Gender', title_fontsize='12')
plt.show()
```

```python
#Proportion of Total Purchase Amount by Category

total_purchase_by_category = df.groupby('Category')['Purchase Amount (USD)'].sum()

# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(total_purchase_by_category, labels=total_purchase_by_category.index, autopct='%1.1f%%', startangle=90)
plt.title('Proportion of Total Purchase Amount by Category')
plt.show()
```
